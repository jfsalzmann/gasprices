```python
# Import some libraries that we will use

import pandas as pd
import matplotlib.pylab as plt
from joffpype import _,pipes

# Then read the file as a csv

df = pd.read_csv("https://raw.githubusercontent.com/jfsalzmann/gasprices/main/data-constr/masters_jointz.csv")
```

```python
# We can take a look at the first couple of samples
df.head()
```

```python
# We can also get a statistical description of the data
df.describe()
```

```python
# How can we set X and y values with our own data?

from datetime import date

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import FunctionTransformer
```

```python
y = df['y']
X = df.drop('y',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=100/1833, shuffle=False)
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X, y, test_size=100/1833, shuffle=True)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import QuantileTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_validate, GridSearchCV


mod = SGDRegressor(random_state=123)
scaler = StandardScaler()
qt = QuantileTransformer(n_quantiles = 100)

pipe = Pipeline(
    [
      ('scaler', scaler),
      ('qt', qt),
      ('model', mod)
    ])

param_grid = {
    'model__alpha': 0.1,#10.0 ** -np.arange(1, 7),
    'model__loss': ['squared_error'],#['squared_error', 'huber', 'epsilon_insensitive'],
    'model__penalty': ['l1'],#['l2', 'l1', 'elasticnet'],
    'model__learning_rate': ['invscaling']#['constant', 'optimal', 'invscaling'],
}
clf = GridSearchCV(pipe, param_grid)

#Fit and predict the pipe, print results
pred = clf.fit(X_train,y_train).predict(X_test)
pred_s = clf.fit(X_train_s,y_train_s).predict(X_test_s)

pred[0:4] |_| print
pred_s[0:4] |_| print
```

```python
clf.best_params_
```

```python
cross_validate(pipe, X_train, y_train, cv = 10, return_estimator=True)
```

```python
# How does the data look after seeing the description?

plt.scatter(pred, y_test, c=X_test["date"]);
```

```python
plt.scatter(pred, pred-y_test, c=X_test["date"]);
```

```python
from sklearn import metrics
import numpy as np

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
print('R squared:', metrics.r2_score(y_test, pred))
```
